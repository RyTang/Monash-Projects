```python
df = pd.read_csv("gdp_country.csv")
print(df)
```

```python
df2 = pd.melt(df, id_vars="Country", value_vars= [
                                    "1999",
                                    "2000",
                                    "2001",
                                    "2002",
                                    "2003",
                                    "2004",
                                    "2005",
                                    "2006",
                                    "2007",
                                    "2008",
                                    "2009",
                                    "2010",
                                    "2012",
                                    "2013",
                                    "2014",
                                    "2015",
                                    "2016",
                                    "2017",
                                    "2018",
                                    "2019",
                                    "2020",
                                    "2021","2022"], var_name='Year', value_name='gdp')
```

```python
df2['Country'] = df2['Country'].astype(str)
df2['Year'] = df2['Year'].astype('int')
df2['gdp'] = df2['gdp'].replace(',', '', regex = True)
df2['gdp'] = df2['gdp'].astype(float)
```

```python
df2
```

```python
df2.to_csv('gdp_clean.csv', index=False)
```

```python
population = pd.read_csv("world_population.csv")
```
